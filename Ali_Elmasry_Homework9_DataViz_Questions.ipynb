### **Question 1**

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns

hd = pd.read_csv("housing_california.csv")
```

#### Histogram of Median Housing Prices:



```python
plt.figure(figsize=(10,6))
plt.hist(hd["median_house_value"], bins= 100)
plt.xlabel("Median Housing Prices")
plt.ylabel("Number of Houses")

plt.show()









```

### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
x = hd["population"]
y = hd["median_house_value"]
plt.scatter(x, y, alpha=0.3)
plt.title('Scatter Plot of Population vs Median House Value')
plt.xlabel('Population')
plt.ylabel('Median House Value')
plt.grid(True)
plt.show()


```

### **Question 3**

#### Box Plot of Median Income:



```python
plt.boxplot(hd['median_income'])
plt.title('Box Plot of Median Income')
plt.ylabel('Median Income')
plt.show()


```



---

### **Question 4**

#### Pair Plot for Select Columns:


```python
columns = ['median_income', 'median_house_value', 'housing_median_age', 'population']
sns.pairplot(hd[columns])
plt.suptitle('Pair Plot of Selected Columns', y=1)
plt.show()










```

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python
counts = hd['ocean_proximity'].value_counts()

counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Ocean Proximity Counts')
plt.xlabel('Ocean Proximity')
plt.ylabel('Count')
plt.show()
```

### **Question 6**

#### Correlation Heatmap:


```python
correlation_matrix = hd.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

```

### **Question 7**

#### Density Plot of Housing Median Age:


```python
sns.kdeplot(hd['housing_median_age'], color='blue')
plt.title('Density Plot of Housing Median Age')
plt.xlabel('Housing Median Age')
plt.ylabel('Density')
plt.grid(axis='y', linestyle='--')
plt.show()

```

### **Question 8**

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python
hd = hd.sort_values(by='housing_median_age')

plt.stackplot(hd['housing_median_age'], hd['total_bedrooms'], hd['total_rooms'], labels=['Total Bedrooms', 'Total Rooms'])
plt.title('Stacked Area Plot of Total Bedrooms and Total Rooms')
plt.xlabel('Housing Median Age')
plt.ylabel('Count')
plt.show()



```

### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
plt.hexbin(hd['longitude'], hd['latitude'], gridsize=50, cmap='viridis')
plt.colorbar(label='Count in Bin')
plt.title('Hexbin Plot of Longitude and Latitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

```

### **Question 10**

#### Facet Grid of Scatter Plots:


```python
columns = ['median_income', 'median_house_value', 'housing_median_age', 'population']

sns.set(style="ticks")
scatter_plot = sns.PairGrid(hd, vars=columns)
scatter_plot.map(plt.scatter)
scatter_plot.add_legend()
plt.suptitle('Facet Grid Scatter Plot of Selected Columns', y=1.02)
plt.show()

```
