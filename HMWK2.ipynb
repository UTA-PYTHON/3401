* Given the equation
$6+2×(5−3^2)+4/2$:

 * * Solve the equation step-by-step, following the PEMDAS order, and comment
 each step in the Jupyter cell.

```python
# First, the parentheses are done

# 3 is squared:  6 + 2*(5-9) + 4/2
# Subtract 9 from 5: 6 + 2*(-4) + 4/2

# Next, all multiplication or division is done left to right

# Multiply 2 times -4: 6 + -8 + 4/2
# Divide 4 by 2: 6 + -8 + 2

# Finally, we add or subtract left to right

# Add 6 + -8: -2 + 2
# Add -2 and 2: 0

# 0 is our final answer
```
