```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/housing_california.csv")
```

```python
# Question 1: Histogram of Median Housing Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['median_house_value'], bins=30, kde=True)
plt.title('Histogram of Median Housing Prices')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Question 2: Scatter Plot of Population vs. Median House Value
plt.figure(figsize=(10, 6))
sns.scatterplot(x='population', y='median_house_value', data=df)
plt.title('Scatter Plot of Population vs. Median House Value')
plt.xlabel('Population')
plt.ylabel('Median House Value')
plt.show()
```

```python
# Question 3: Box Plot of Median Income
plt.figure(figsize=(10, 6))
sns.boxplot(x='median_income', data=df)
plt.title('Box Plot of Median Income')
plt.xlabel('Median Income')
plt.show()
```

```python
# Question 4: Pair Plot for Select Columns
selected_columns = ['median_house_value', 'median_income', 'population', 'housing_median_age']
sns.pairplot(df[selected_columns])
plt.suptitle('Pair Plot for Select Columns', y=1.02)
plt.show()
```

```python
# Question 5: Bar Plot of Ocean Proximity Counts
plt.figure(figsize=(10, 6))
sns.countplot(x='ocean_proximity', data=df)
plt.title('Bar Plot of Ocean Proximity Counts')
plt.xlabel('Ocean Proximity')
plt.ylabel('Count')
plt.show()
```

```python
# Question 6: Correlation Heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Question 7: Density Plot of Housing Median Age
plt.figure(figsize=(10, 6))
sns.kdeplot(df['housing_median_age'], fill=True)
plt.title('Density Plot of Housing Median Age')
plt.xlabel('Housing Median Age')
plt.ylabel('Density')
plt.show()
```

```python
# Question 8: Stacked Area Plot of Total Rooms and Bedrooms
df['cumulative_total_rooms'] = df['total_rooms'].cumsum()
df['cumulative_bedrooms'] = df['total_bedrooms'].cumsum()

plt.figure(figsize=(10, 6))
plt.stackplot(df['housing_median_age'], df['cumulative_total_rooms'], df['cumulative_bedrooms'], labels=['Total Rooms', 'Total Bedrooms'])
plt.title('Stacked Area Plot of Total Rooms and Bedrooms')
plt.xlabel('Housing Median Age')
plt.ylabel('Cumulative Count')
plt.legend(loc='upper left')
plt.show()
```

```python
# Question 9: Hexbin Plot of Latitude and Longitude
plt.figure(figsize=(10, 6))
sns.jointplot(x='longitude', y='latitude', data=df, kind='hex', gridsize=20)
plt.title('Hexbin Plot of Latitude and Longitude')
plt.show()
```

```python
# Question 10: Facet Grid of Scatter Plots
sns.set(style="ticks")
g = sns.FacetGrid(df, col="ocean_proximity", hue="ocean_proximity", col_wrap=4)
g.map(plt.scatter, "population", "median_house_value", alpha=.7)
g.add_legend()
plt.suptitle('Facet Grid of Scatter Plots', y=1.02)
plt.show()
```
