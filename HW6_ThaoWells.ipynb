```python
#1.

class Vehicle:
    def __init__(self, make, model, year, fuel):
        self.make = make
        self.model = model
        self.year = year
        self.fuel = fuel

    def info(self):
        print(f"Make: {self.make}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")
        print(f"Fuel Efficiency: {self.fuel} mpg")



class Car(Vehicle):
    def __init__(self, make, model, year, fuel, doors):
        super().__init__(make, model, year, fuel)
        self.doornum = doors

    def info(self):
        super().info()
        print(f"Number of doors: {self.doors} doors")

class Bicycle(Vehicle):
    def __init__(self, make, model, year, fuel, type):
        super().__init__(make, model, year, fuel)
        self.type = type

    def info(self):
        super().info()
        print(f"Bicycle Type: {self.type}")
```

```python
#2.
class Car(Vehicle):
    def __init__(self, make, model, year, fuel, num_doors, is_sedan):
        super().__init__(make, model, year, fuel)
        self.num_doors = num_doors
        self.is_sedan = is_sedan

    def display_car_info(self):
        super().info()
        print(f"Number of Doors: {self.num_doors}")
        sedan_status = "Yes" if self.is_sedan else "No"
        print(f"Is this vehicle a Sedan? {sedan_status}")
```

```python
#3.
class Motorcycle(Vehicle):
    def __init__(self, make, model, year, fuel, top_speed, engine_size):
        super().__init__(make, model, year, fuel)
        self.top_speed = top_speed
        self.engine_size = engine_size

    def display_moto(self):
        super().info()
        print(f"Top Speed: {self.top_speed} mph")
        print(f"Engine Size: {self.engine_size} cc")


car_instance = Car("Toyota", "Camry", 2023, 30.5, 4, True)
motorcycle_instance = Motorcycle("Honda", "CBR1000RR", 2023, 40.0, 186, 999)


print("Car Information:")
car_instance.display_car_info()

print("\nMotorcycle Information:")
motorcycle_instance.display_moto()
```
