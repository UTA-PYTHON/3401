### **Question 1**

#### Histogram of Median Housing Prices:



```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns

df = pd.read_csv("housing_california.csv")

plt.figure(figsize=(10,6))
plt.hist(df["median_house_value"], bins= 100)
plt.xlabel("Median Housing Prices")
plt.ylabel("Number of Houses")
plt.show()








```

### **Question 2**

#### Scatter Plot of Population vs. Median House Value:



```python
x = df["population"]
y = df["median_house_value"]
plt.scatter(x, y, alpha=0.3)
plt.title('Population vs Median House Value')
plt.xlabel('Population')
plt.ylabel('Median House Value')
plt.grid(True)
plt.show()










```

### **Question 3**

#### Box Plot of Median Income:



```python
plt.boxplot(df['median_income'])
plt.title('Box Plot of Median Income')
plt.ylabel('Median Income')
plt.show()
```

### **Question 4**

#### Pair Plot for Select Columns:


```python
column = ['median_income', 'median_house_value', 'housing_median_age', 'population']
sns.pairplot(df[column])
plt.suptitle('Pair Plot of Selected Columns', y=1)
plt.show()










```

### **Question 5**

#### Bar Plot of Ocean Proximity Counts:


```python
counts = df['ocean_proximity'].value_counts()
counts.plot(kind='bar', edgecolor='black')

plt.title('Ocean Proximity Counts')
plt.xlabel('Ocean Proximity')
plt.ylabel('Counts')
plt.show()








```

### **Question 6**

#### Correlation Heatmap:


```python
correlation_matrix = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()











```

### **Question 7**

#### Density Plot of Housing Median Age:


```python

sns.kdeplot(df['housing_median_age'], color='blue')
plt.title('Density Plot of Housing Median Age')
plt.xlabel('Housing Median Age')
plt.ylabel('Density')
plt.grid(axis='y', linestyle='--')
plt.show()













```



---

### **Question 8**

#### Stacked Area Plot of Total Rooms and Bedrooms:


```python

df = df.sort_values(by='housing_median_age')

plt.stackplot(df['housing_median_age'], df['total_bedrooms'], df['total_rooms'], labels=['Total Bedrooms', 'Total Rooms'])
plt.title('Total Bedrooms and Rooms')
plt.xlabel('Housing Median Age')
plt.ylabel('Count')
plt.show()
















```

### **Question 9**

#### Hexbin Plot of Latitude and Longitude:


```python
plt.hexbin(df['longitude'], df['latitude'], gridsize=40, cmap='viridis')
plt.colorbar(label='Count in Bin')
plt.title('Longitude and Latitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()











```

### **Question 10**

#### Facet Grid of Scatter Plots:


```python
column = ['median_income', 'median_house_value', 'housing_median_age', 'population']
sns.set(style="ticks")

scatter_plot = sns.PairGrid(df, vars=column)
scatter_plot.map(plt.scatter)

scatter_plot.add_legend()
plt.suptitle('Facet Grid of Scatter Plots for median income, house value, median age, and population', y=1.5)
plt.show()











```
